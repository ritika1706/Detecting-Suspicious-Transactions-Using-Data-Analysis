```python
# extract time features
df['Hour'] = df['TransactionDate'].dt.hour
df['Day'] = df['TransactionDate'].dt.day
df['Month'] = df['TransactionDate'].dt.month

# transaction count per account
df['Txn_Count'] = df.groupby('AccountID')['TransactionID'].transform('count')

# avg transaction per account
df['Avg_Amount_Per_Account'] = df.groupby('AccountID')['TransactionAmount'].transform('mean')
```

```python
df['Prev_Device'] = df.groupby('AccountID')['DeviceID'].shift(1)
df['Device_Changed'] = df['DeviceID'] != df['Prev_Device']

df['Prev_Location'] = df.groupby('AccountID')['Location'].shift(1)
df['Location_Changed'] = df['Location'] != df['Prev_Location']
df.head()
```

 # 🚀 What a Data Analyst would do next
✅ 1. Create a risk profile (VERY IMPORTANT)

Combine all your features:

```python
# risk_df = customer summary 
risk_df = df.groupby('AccountID').agg({
    'DeviceID': 'nunique',
    'LoginAttempts': 'max',
    'TransactionID': 'count',
    'TransactionAmount': 'mean',
    'Device_Changed': 'sum',
    'Location_Changed': 'sum'
}).rename(columns={
    'DeviceID': 'Device_Count',
    'LoginAttempts': 'Max_LoginAttempts',
    'TransactionID': 'Txn_Count',
    'TransactionAmount': 'Avg_Amount'
})
risk_df.head()
```

```python
suspicious = risk_df[
    (risk_df['Device_Count'] > 5) |
    (risk_df['Txn_Count'] > 200) |
    (risk_df['Max_LoginAttempts'] > 3) |
    (risk_df['Device_Changed']>100)|
      (risk_df['Location_Changed']>100)
       ]

suspicious
```

```python
df['Date'] = df['TransactionDate'].dt.date
```

# per-day behavior


```python
daily_df = df.groupby(['AccountID', 'Date']).agg({
   'DeviceID': 'nunique',
    'LoginAttempts': 'max',
    'TransactionID': 'count',
    'TransactionAmount': 'mean',
    'Device_Changed': 'sum',
    'Location_Changed': 'sum'
}).reset_index()
daily_df.describe()
```

```python
suspicious_daily = daily_df[
   (daily_df['DeviceID'] > 5) |
    (daily_df['TransactionID'] > 10) |
    (daily_df['LoginAttempts'] > 3) |
    (daily_df['Device_Changed']>5)|
      (daily_df['Location_Changed']>5)
]
suspicious_daily
```

# no . of suspicious accounts

```python
suspicious_accounts_daily = suspicious_daily['AccountID'].unique()
len(suspicious_accounts_daily)
```

# complete fraud detection

👤 Case 1:

User always behaves strangely
👉 risky account
risk_df = “Who is suspicious”
👤 Case 2:

User normal for months

Suddenly 100 transactions in 1 day

👉 fraud event

daily_df = when something suspicious happened

```python
suspicious_accounts_overall = suspicious.index

suspicious_accounts_daily = suspicious_daily['AccountID'].unique()
```

```python
# “Give me ALL users who are suspicious in ANY way”
combined_accounts = set(suspicious_accounts_overall).union(set(suspicious_accounts_daily))
len(combined_accounts)
```

```python
final_df = risk_df.copy()
final_df['Fraud_Flag'] = final_df.index.isin(combined_accounts).astype(int)
final_df.head()
```
